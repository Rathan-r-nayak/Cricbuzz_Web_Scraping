```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
head={"User-agent":'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
ws=[]
for i in range(0,15000):
    url=f"https://www.cricbuzz.com/profiles/{i}"
    ws.append(requests.get(url,headers=head).text)
type(ws)
```

```python
soup=BeautifulSoup(str(ws),'lxml')
# print(soup.find_all('h1'))
```

```python
content=soup.find_all('div',id='playerProfile')
print(content[0].find('h1').text)
```

```python
dict={
    "Name":pname,
    "Country":pcountry,
    "Test Matches":test_matches,
    "Test Runs":test_runs,
    "Test Balls":test_balls,
    "Test Avg":test_avg,
    "Test SR":test_sr,
    "Odi Matches":odi_matches,
    "Odi Runs":odi_runs,
    "Odi Balls":odi_balls,
    "Odi Avg":odi_avg,
    "odi SR":odi_sr,
    "T20 Match":t_matches,
    "T20 Runs":t_runs,
    "T20 Balls":t_balls,
    "T20 Avg":t_avg,
    "T20 SR":t_sr
}
```
